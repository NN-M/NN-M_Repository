```python
from pykakasi import kakasi

text1 = "汝、文学と科学の力を信じよ"
text2 = '情報分析して改善アクション'

kakasi = kakasi()

kakasi.setMode('J', 'K')
kakasi.setMode('H', 'K')

conv = kakasi.getConverter()

word1 = conv.do(text1).replace('、', '')
word2 = conv.do(text2)
```

```python
word2
```

```python
start = 0
end = len(word1)
left = 0
right = len(word2)
count = 0

match1_index=[]
match2_index=[]
answer = []

while left < right:
    match1 = -1
    while match1 == -1:
        match1 = word1.find(word2[left], start)
        left += 1

    match2 = -1
    while match2 == -1:
        match2 = word1.rfind(word2[right - 1], start, end)
        right -= 1

    if match1 < match2:
        answer[count:count] = [word1[match1], word1[match2]]
        match1_index.append(match1)
        match2_index.append(match2)
        start = match1 + 1
        end = match2
        count += 1
        
    elif match1 == match2:
        answer.insert(count, word1[match1])
        break

    elif match1 == min(match1, min(match2_index)):
        left += 1
        
    elif match2 == max(match2, max(match1_index)):
        right += 1

answer
```
